  INTRODUCTION
</h2>

The Bank Marketing Project focuses on analyzing a dataset collected from a Portuguese banking institution’s marketing campaign. The main objective of the campaign was to persuade potential clients to subscribe to a term deposit product. This dataset contains information on client demographics, their previous interactions with the bank, and the outcomes of marketing campaigns.

The goal of this project is to build a robust machine learning model that can predict whether a client will subscribe to a term deposit based on their personal and interaction details. By accurately identifying potential subscribers, the bank can optimize its marketing strategy, reduce campaign costs, and improve overall efficiency.

This project involves several steps including data cleaning, exploratory data analysis (EDA), feature engineering, encoding, model training, and performance evaluation using advanced classification algorithms such as XGBoost and CatBoost. The ultimate aim is to achieve a highly accurate and interpretable model that helps the bank make data-driven marketing decisions.

```python
# Create interaction term
df['campaign_previous'] = df['campaign'] * df['previous']
```

```python
df['campaign_previous']
```

```python
# Age Grouping :
```

```python
def age_group(age):
    if age < 30:
        return 'Young'
    elif age < 60:
        return 'Middle-aged'
    else:
        return 'Senior'
df['age_group'] = df['age'].apply(age_group)
```

```python
df['age_group']
```

```python
# Balance Level Binning :
```

```python
def balance_level(balance):
    if balance < 0:
        return 'Debt'
    elif balance < 1000:
        return 'Low'
    elif balance < 5000:
        return 'Medium'
    else:
        return 'High'
df['balance_level'] = df['balance'].apply(balance_level)
```

```python
df['balance_level'] 
```

```python
# Checking the customer has any loans
```

```python
def check_any_loan(row):
    if row['housing'] == 'yes' or row['loan'] == 'yes':
        return 1
    else:
        return 0

df['has_any_loan'] = df.apply(check_any_loan, axis=1)
```
