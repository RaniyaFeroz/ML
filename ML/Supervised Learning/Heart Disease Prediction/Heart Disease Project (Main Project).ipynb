  ABSTRACT
</h2>

Heart disease is one of the leading causes of death worldwide, and early detection plays a vital role in reducing its impact. This project focuses on developing a machine learning model to predict the likelihood of heart disease based on clinical and demographic features. Machine Learning algorithms like  Logistic Regression and Support Vector Machine (SVM) were tested and evaluated. Among them, Logistic Regression achieved the highest accuracy of 92.59%, showing strong predictive performance. The model can assist healthcare professionals in identifying high-risk patients, enabling timely diagnosis and prevention.

  INTRODUCTION
</h2>

Heart disease prediction has become an essential application of data science and machine learning in healthcare. Traditional diagnosis relies on clinical tests and expert interpretation, which can be time-consuming. Machine learning offers an efficient alternative by analyzing patient data to identify patterns associated with heart disease risk.

The goal of this project is to build and evaluate predictive models that can accurately classify patients as having or not having heart disease. The models are trained on key features such as age, sex, chest pain type, cholesterol level, resting blood pressure, fasting blood sugar, ECG results, maximum heart rate, exercise-induced angina, oldpeak, slope of the ST segment, major vessels and thalassemia type.

  About the Dataset
</h2>

The dataset used in this project is derived from the Cleveland Heart Disease Dataset, a well-known benchmark in cardiovascular research. It contains 270 patient records and 14 attributes, including both categorical and numerical features related to patient demographics, clinical test results, and cardiac health indicators.

The features used in this study are : age,  sex, chest_pain_type,  resting_blood_pressure, serum_cholestoral,  fasting_blood_sugar, resting_electrocardiographic_results, max_heart_rate,  exercise_induced_angina, oldpeak,  st_segment,  major_vessels,  thal, and heart_disease.

The target variable, heart_disease, indicates whether a patient is diagnosed with heart disease (1) or not (0). The dataset provides a balanced representation of both classes, allowing effective model training and evaluation.

```python
outlier_cols = ["chest_pain_type","resting_blood_pressure","serum_cholestoral","fasting_blood_sugar","max_heart_rate","oldpeak","major_vessels"]
```

```python
for col in outlier_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = np.clip(df[col], lower, upper)  
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show() 
```

  STATISTICAL T-TEST
</h2>

```python
from scipy.stats import ttest_ind

# Split age by heart disease groups
age_no_hd = df[df['heart_disease'] == 0]['age']   # people without heart disease
age_yes_hd = df[df['heart_disease'] == 1]['age']  # people with heart disease

# Perform independent t-test
t_stat, p_value = ttest_ind(age_no_hd, age_yes_hd)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
# Interpretation
if p_value < 0.05:
    print("Result: Statistically significant difference in age between heart disease groups")
else:
    print("Result: No statistically significant difference in age between heart disease groups")
```

* ### The independent t-test comparing age between people with and without heart disease shows a statistically significant difference (p < 0.001). This indicates that age is strongly associated with heart disease in this dataset.
* ### Specifically, individuals with heart disease tend to be older on average than those without, suggesting that age is an important risk factor to consider in predicting heart disease.
