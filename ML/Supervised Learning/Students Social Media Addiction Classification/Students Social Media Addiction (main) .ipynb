  INTRODUCTION
</h2>

The dataset titled “Students Social Media Addiction” aims to explore the relationship between social media usage and various aspects of student life. It contains detailed information on students from diverse academic levels and countries, capturing their daily social media habits, preferred platforms, sleep patterns, academic performance, mental health, and relationship status. By analyzing variables such as average daily usage hours, addiction scores, and conflicts arising from social media, this dataset helps identify patterns and potential impacts of excessive social media use. It serves as a valuable resource for understanding how digital behavior influences both academic and personal well-being among students in the modern era.

  About the Dataset
</h2>

This dataset contains information collected from students across various academic levels and countries, focusing on their social media habits and its effects on their daily lives. It includes 13 key attributes for each student, covering demographic details, social media usage patterns, mental health indicators, and relationship dynamics.

The primary objective of the dataset is to analyze how social media usage impacts students’ academic performance, mental health, and sleep patterns, while also exploring addiction levels and social conflicts stemming from digital interactions.

">
  Statistical Significance of Gender-Based Addiction Score Differences
</h3>

```python
from scipy.stats import ttest_ind

female_scores = df[df['Gender'] == 'Female']['Addicted_Score']
male_scores = df[df['Gender'] == 'Male']['Addicted_Score']

t_stat, p_val = ttest_ind(female_scores, male_scores)
print(p_val)
```

```python
if p_val < 0.05:
    print("✅ The difference in addiction scores between males and females is statistically significant.")
else:
    print("❌ The difference in addiction scores between males and females is not statistically significant.")
```

## Distribution of Social Media Platform Usage

```python
plt.figure(figsize=(14, 7))
platform_counts = df['Most_Used_Platform'].value_counts()
sns.barplot(x=platform_counts.index, y=platform_counts.values, palette='viridis')
plt.title('Most Used Social Media Platforms', fontsize=14)
plt.xlabel('Platform', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.show()
```

## Insight :
#### Instagram is the most used platform, followed by TikTok and Facebook, indicating they are the dominant platforms among users. Others like WhatsApp and Twitter are moderately used, while platforms like LinkedIn and Snapchat have lower engagement.

## Proportion of Users by Education Level

```python
# Academic level pie chart
level_counts = df['Academic_Level'].value_counts()

plt.figure(figsize=(7, 7))
plt.pie(level_counts, labels=level_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Academic Level Distribution')
plt.axis('equal')
plt.show()
```

## Insight :
#### The majority of respondents are undergraduates (50.1%), followed closely by graduates (46%). Only a small portion (3.8%) are high school students. This indicates that social media usage insights are primarily driven by higher education students.

## User Frequency by Daily Social Media Usage Hours

```python
# Average Daily Usage Hours
plt.figure(figsize=(12, 6))
sns.histplot(df['Avg_Daily_Usage_Hours'], kde=True, bins=20)
plt.title('Average Daily Social Media Usage Hours', fontsize=14)
plt.xlabel('Hours', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.axvline(df['Avg_Daily_Usage_Hours'].mean(), color='red', linestyle='--', label=f'Mean: {df["Avg_Daily_Usage_Hours"].mean():.2f} hours')
plt.legend()
plt.show()
```

## Insight :
#### The distribution shows that most respondents spend between 4 to 6 hours daily on social media, with a mean usage of 4.92 hours. The curve is slightly right-skewed, indicating that while a majority spend a moderate amount of time, a smaller group uses social media for longer hours. This suggests high engagement levels, especially among younger users.

## Proportion of Respondents by Relationship Category

## Correlation Heatmap of Numerical Features

```python
# Select relevant numerical columns for correlation
numerical_cols = ['Age', 'Avg_Daily_Usage_Hours', 'Sleep_Hours_Per_Night',
                  'Mental_Health_Score', 'Conflicts_Over_Social_Media', 'Addicted_Score']

# Compute Pearson correlation matrix
correlation_matrix = df[numerical_cols].corr(method='pearson')

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, linecolor='gray')
plt.title("Correlation Heatmap")
plt.xticks(rotation=30)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

## Insight :
    •	People who use social media more are more likely to be addicted. → Correlation = 0.83
	•	People who use social media more often have more fights or problems because of it. → Correlation = 0.80
	•	People who are more addicted to social media also face more conflicts. → Correlation = 0.93
	•	More social media use is linked with poorer mental health. → Correlation = -0.80
	•	Being more addicted to social media is linked with worse mental health. → Correlation = -0.95
	•	More conflicts over social media are linked to worse mental health. → Correlation = -0.89
	•	People who use social media more usually sleep less. → Correlation = -0.79
	•	People who are addicted to social media sleep less. → Correlation = -0.76
	•	People who sleep more usually have better mental health. → Correlation = 0.71
	•	Age does not have a strong connection with any of these things. → Correlations are small (between -0.18 and 0.16)

## More social media = more addiction, more conflicts, less sleep, and worse mental health. Age doesn’t matter much.
